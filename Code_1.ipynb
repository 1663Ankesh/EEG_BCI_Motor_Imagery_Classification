```python
import numpy as np

# Initialize an empty list to collect labels
flat_labels = []

for trial in tr:
    y_field = trial['y']
    
    # Check if 'y' is non-empty
    if y_field.size > 0:
        # Flatten and extend the list if y_field is a non-empty array
        for item in y_field:
            if isinstance(item, np.ndarray):
                flat_labels.extend(item.flatten())
            else:
                flat_labels.append(item)

# Convert list of arrays to a flat list of scalars
# Ensure to handle cases where flat_labels might already be flattened
flat_labels = [item for sublist in flat_labels if isinstance(sublist, np.ndarray) for item in sublist]

# Convert to a NumPy array
labels = np.array(flat_labels)

# Print the number of labels and the first few for verification
print(f"Number of Labels Extracted: {len(labels)}")
print("First few labels:", labels[:10])
```

```python
# Preprocess EEG data if needed
eeg_data = np.expand_dims(eeg_data, axis=-1)  # Add a dimension for channels if necessary
```

```python
encoder = LabelEncoder()
labels = encoder.fit_transform(labels)
labels = to_categorical(labels)
```

```python
eeg_data = []  # Initialize empty list to collect EEG data

for trial in eeg_data_struct:
    X_field = trial[0][0]['X']
    if X_field.size > 0:
        eeg_data.extend(X_field)

# Convert to NumPy array
eeg_data = np.array(eeg_data)
```

```python
print(f"Shape of eeg_data: {eeg_data.shape}")
print(f"Shape of labels: {labels.shape}")
```

```python
# Flatten eeg_data and reshape
eeg_data_flat = eeg_data.flatten()
num_samples = 288
timepoints_per_sample = 25614
num_features = 25

# Check if reshaping is possible
expected_elements = num_samples * timepoints_per_sample * num_features
if eeg_data_flat.size == expected_elements:
    eeg_data_reshaped = eeg_data_flat.reshape((num_samples, timepoints_per_sample, num_features))
    print(f"Shape of reshaped eeg_data: {eeg_data_reshaped.shape}")
else:
    print(f"Cannot reshape eeg_data. Total elements mismatch. Got {eeg_data_flat.size}, expected {expected_elements}.")
```

```python
# Convert labels to numpy array if not already
labels = np.array(labels)

# Ensure labels shape matches the number of samples
if labels.shape[0] == num_samples:
    print(f"Shape of labels: {labels.shape}")
else:
    print(f"Labels shape mismatch. Got {labels.shape[0]}, expected {num_samples}.")
```

```python
# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(eeg_data, labels, test_size=0.2, random_state=42)
```
