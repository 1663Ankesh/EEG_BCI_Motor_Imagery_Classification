```python
print(f"Training Data Shape: {X_train.shape}")
print(f"Test Data Shape: {X_test.shape}")
print(f"Training Labels Shape: {y_train.shape}")
```

```python
def build_multiscale_cnn(input_shape):
    input_layer = layers.Input(shape=input_shape)

    branch1 = layers.Conv1D(16, 3, activation='relu', padding='same')(input_layer)
    branch1 = layers.Conv1D(32, 3, activation='relu', padding='same')(branch1)
    branch1 = layers.MaxPooling1D(2)(branch1)

    branch2 = layers.Conv1D(16, 5, activation='relu', padding='same')(input_layer)
    branch2 = layers.Conv1D(32, 5, activation='relu', padding='same')(branch2)
    branch2 = layers.MaxPooling1D(2)(branch2)

    branch3 = layers.Conv1D(16, 7, activation='relu', padding='same')(input_layer)
    branch3 = layers.Conv1D(32, 7, activation='relu', padding='same')(branch3)
    branch3 = layers.MaxPooling1D(2)(branch3)

    merged = layers.concatenate([branch1, branch2, branch3], axis=-1)
    merged = layers.Flatten()(merged)
    
    fc = layers.Dense(64, activation='relu')(merged)
    output = layers.Dense(y_train.shape[1], activation='softmax')(fc)

    model = models.Model(inputs=input_layer, outputs=output)
    return model
```

```python
X_train_reshaped = X_train.reshape(X_train.shape[0], X_train.shape[1], 1) 
X_test_reshaped = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)  

print(f"X_train reshaped: {X_train_reshaped.shape}") 
print(f"X_test reshaped: {X_test_reshaped.shape}")  
```

```python
print(f"Shape of y_train: {y_train.shape}")
```

```python
num_classes = len(np.unique(y_train))  
y_train = to_categorical(y_train, num_classes=num_classes)
y_test = to_categorical(y_test, num_classes=num_classes) 

print(f"Number of classes: {num_classes}")
print(f"Shape of y_train after one-hot encoding: {y_train.shape}")
print(f"Shape of y_test after one-hot encoding: {y_test.shape}")
```

```python
input_shape = (X_train_reshaped.shape[1], X_train_reshaped.shape[2])  
input_shape
```

```python
model = build_multiscale_cnn(input_shape)
```

```python
epc=20
bs=32
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train_reshaped, y_train, validation_data=(X_test_reshaped, y_test), epochs=epc, batch_size=bs)

loss, accuracy = model.evaluate(X_test_reshaped, y_test)
print(f'Test accuracy: {accuracy}')
```
